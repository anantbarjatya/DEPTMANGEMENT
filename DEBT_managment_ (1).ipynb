```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np

# Sample data (replace this with your actual df_combined if already defined)
df_combined = pd.DataFrame({
    'Year': [1950, 1951, 1952, 1953, 1954],
    'central': [29.56, 25.54, 25.42, 25.35, 24.77],
    'general': [64.83, 57.93, 54.43, 53.09, 51.52],
    'household': [21.29, 19.47, 20.20, 21.67, 23.25],
    'corporate': [26.04, 26.08, 28.47, 29.42, 31.68],
    'private': [26.02, 29.46, 29.37, 34.25, 36.88]
})

# Set 'Year' as index
df_combined.set_index('Year', inplace=True)

# Function for debt forecasting
def forecast_debt(df_combined, debt_type='central'):
    X = df_combined.index.values.reshape(-1, 1)  # Year as the feature
    y = df_combined[debt_type].values           # Corrected line

    model = LinearRegression()
    model.fit(X, y)

    future_year = np.array([[2025]])
    future_debt = model.predict(future_year)

    return future_debt[0]

# Forecast for central debt in 2025
future_central_debt = forecast_debt(df_combined, debt_type='central')
print(f"Forecasted central debt for 2025: {future_central_debt:.2f}")


```

```python
debt_types = ['central', 'general', 'household', 'corporate', 'private']

for debt in debt_types:
    predicted = forecast_debt(df_combined, debt)
    print(f"Forecasted {debt} debt for 2025: {predicted:.2f}")
```
